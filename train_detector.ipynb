# Label format

```python
import json
import numpy as np
def generate_txt_with_unified3dboxdefinition(path:str):
    out = ""
    with open(path, 'r') as f:
        a = f.read()
        b = json.loads(a)
        print(b[0])
        for o in b:
            x = o["location"]["x"]
            y = o["location"]["y"]
            z = o["location"]["z"]
            dx = o["extent"]["x"]
            dy = o["extent"]["y"]
            dz = o["extent"]["z"]
            heading_angle = np.radians(o["rotation"]["yaw"])
            category_name = "Vehicle"
            if "pedestrian" in o["bp_id"]:
                category_name = "Pedestrian"
            elif "bike" in o["bp_id"]:
                category_name = "Cyclist"
            out += "{x} {y} {z} {dx} {dy} {dz} {heading_angle} {category_name}\n".format(x=x, y=y, 
                                                                                         z=z, dx=dx, 
                                                                                         dy=dy, dz=dz, 
                                                                                         heading_angle=heading_angle, 
                                                                                         category_name=category_name)
        return out
generate_txt_with_unified3dboxdefinition("/home/filo/Downloads/BBOX_LABELS/Town01_Opt_ClearNoon_5300274882951739.json")
```
